```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python


data_path = r'C:\Users\X280\Downloads'
df = pd.read_csv(data_path + r'\datavarsmall.csv', header=0)

```

```python
df
```

```python
df.describe()
```

```python
g = sns.FacetGrid(df, col="FS_SIZE", col_wrap=3)
g.map(sns.boxplot, "File_name", "f1-score")
```

```python
ax=sns.lineplot(data=df,x='Selected_features',y='precision',label='Precision')
ax2=sns.lineplot(data=df,x='Selected_features',y='recall',label='Recall')
plt.ylabel("Precison & Recall ")
```

```python
sns.lineplot(data=df, x='FS_SIZE', y='precision',label='Precision')
sns.lineplot(data=df, x='FS_SIZE', y='recall',label='Recall')
plt.ylabel("Precision & Recall")
```

```python
sns.lineplot(data=df, x='Lags', y='precision', label='PrÃ©cision')
sns.lineplot(data=df, x='Lags', y='recall', label='Rappel')
plt.ylabel("Precision & Recall")
```

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=df, kind="bar",
    x="Lags", y="precision", hue="File_name",
    errorbar="sd",palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("Lags", "Precision")
g.legend.set_title("")
```

```python
sns.set_theme(style="whitegrid")
h = sns.catplot(
    data=df, kind="bar",
    x="Lags", y="recall", hue="File_name",
    errorbar="sd",palette="dark", alpha=.6, height=6
)
h.despine(left=True)
h.set_axis_labels("Lags", "Recall")
h.legend.set_title("")
```

```python
sns.set_theme(style="whitegrid")
h = sns.catplot(
    data=df, kind="bar",
    x="Lags", y="recall", hue="target",
    errorbar="sd",palette="dark", alpha=.6, height=6
)
h.despine(left=True)
h.set_axis_labels("Lags", "Recall")
h.legend.set_title("")
```

```python
sns.set_theme(style="whitegrid")
h = sns.catplot(
    data=df, kind="bar",
    x="Lags", y="precision", hue="target",
    errorbar="sd",palette="dark", alpha=.6, height=6
)
h.despine(left=True)
h.set_axis_labels("Lags", "Precision")
h.legend.set_title("")
```
